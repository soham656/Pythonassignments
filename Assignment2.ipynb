```python
from tkinter import *

window=Tk()

e1_value=StringVar()
e1=Entry(window,textvariable=e1_value)
e1.grid(row=0,column=0,padx=10,pady=10)

e2_value=StringVar() # StringVar() is a function by which tkinter moniters the string and updates it
e2=Entry(window,textvariable=e2_value) #textvariable means the input which the user enters basically same as input function
e2.grid(row=0,column=1,padx=10,pady=10)

t1=Text(window,height=1)
t1.grid(row=1,column=0,columnspan=2,padx=10,pady=10)

def addn():
    ans=float(e1_value.get())+float(e2_value.get())
    t1.delete("1.0",END)
    t1.insert(END,ans)

def sub():
    ans1=float(e1_value.get())-float(e2_value.get())
    t1.delete("1.0",END)
    t1.insert(END,ans1)

def mul():
    ans2=float(e1_value.get())*float(e2_value.get())
    t1.delete("1.0",END)
    t1.insert(END,ans2)

def div():
    ans3=float(e1_value.get())/float(e2_value.get())
    t1.delete("1.0",END)
    t1.insert(END,ans3)

def mod():
    ans4=float(e1_value.get())%float(e2_value.get())
    t1.delete("1.0",END)
    t1.insert(END,ans4)

def clear():
    e1.delete(0,END) # 0th chara and then till the end
    e2.delete(0,END)
    t1.delete("1.0",END) # 1st line and from the 0th chara till th eend



b1=Button(window,text="+",width=10,command=addn)
b1.grid(row=2,column=0)

b2=Button(window,text="-",width=10,command=sub)
b2.grid(row=2,column=1)

b3=Button(window,text="*",width=10,command=mul)
b3.grid(row=3,column=0)

b4=Button(window,text="/",width=10,command=div)
b4.grid(row=3,column=1)

b5=Button(window,text="%",width=10,command=mod)
b5.grid(row=4,column=0)

b6=Button(window,text="C",width=10,command=clear)
b6.grid(row=4,column=1)


window.mainloop()
#C to clear input and output data.
```
